/* Q1: Some of the facilities charge a fee to members, but some do not.
Write a SQL query to produce a list of the names of the facilities that do. */

```python
rs = con.execute("Select name, membercost From Facilities Where membercost > 0")
df_facilities_query = pd.DataFrame(rs.fetchall())
df_facilities_query.columns = ['name', 'member cost']
df_facilities_query
```

/* Q2: How many facilities do not charge a fee to members? */

```python
rs = con.execute("Select name, membercost From Facilities Where membercost == 0")
df_facilities_query = pd.DataFrame(rs.fetchall())
df_facilities_query.columns = ['name', 'member cost']
df_facilities_query
```

/* Q3: Write an SQL query to show a list of facilities that charge a fee to members,
where the fee is less than 20% of the facility's monthly maintenance cost.
Return the facid, facility name, member cost, and monthly maintenance of the
facilities in question. */

```python
rs = con.execute("Select facid, name, membercost, monthlymaintenance From Facilities Where membercost < (.20 * monthlymaintenance)")
df_facilities_query = pd.DataFrame(rs.fetchall())
df_facilities_query.drop(0, axis=1, inplace=True)
df_facilities_query.columns = ['name', 'membercost', 'monthly maintenance']
df_facilities_query
```

/* Q4: Write an SQL query to retrieve the details of facilities with ID 1 and 5.
Try writing the query without using the OR operator. */

```python
rs = con.execute("Select * from Facilities Where facid IN (1, 5)")
df_facilities_query = pd.DataFrame(rs.fetchall())
df_facilities_query.columns = rs.keys()
df_facilities_query
```

/* Q5: Produce a list of facilities, with each labelled as
'cheap' or 'expensive', depending on if their monthly maintenance cost is
more than $100. Return the name and monthly maintenance of the facilities
in question. */

```python
rs = con.execute("Select name, monthlymaintenance, Case When monthlymaintenance > 100 Then 'expensive' Else 'cheap' end as cheap_costly from Facilities")
df_facilities_query = pd.DataFrame(rs.fetchall())
df_facilities_query.columns = ["name", "monthlymaintenance", "cheap/expensive"]
df_facilities_query
```

/* Q6: You'd like to get the first and last name of the last member(s)
who signed up. Try not to use the LIMIT clause for your solution. */

```python
rs = con.execute("SELECT firstname|| ' ' ||surname AS Name From members Where joindate >= '2012-09-19 11:32:45'")
df_members = pd.DataFrame(rs.fetchall())
df_members.columns = ['name']
df_members
```

/* Q7: Produce a list of all members who have used a tennis court.
Include in your output the name of the court, and the name of the member
formatted as a single column. Ensure no duplicate data, and order by
the member name. */

```python
rs = con.execute("SELECT DISTINCT firstname || ' ' || surname AS member, name AS facility FROM members INNER JOIN bookings ON (members.memid = bookings.memid) INNER JOIN facilities ON (bookings.facid = facilities.facid) WHERE facilities.name LIKE 'Tennis%' ORDER BY member, facility")
df_members_courts = pd.DataFrame(rs.fetchall())
df_members_courts.columns = ['name', 'court']
df_members_courts
```
